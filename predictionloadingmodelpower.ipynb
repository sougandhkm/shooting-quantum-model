```python
from tensorflow import keras
import numpy as np
```

```python
model = keras.models.load_model('modelpower200.h5')
```

```python
dataset=np.loadtxt("datapowertest.csv", delimiter="\t")                     #loading the test data file
x=dataset[:,0:3]
y=dataset[:,3]
yn = y.reshape((-1, 1))
```

```python
ynew=model.predict(x)                                                    #predicting data
```

```python
error=ynew-yn
```

```python
import matplotlib.pyplot as plt
plt.plot(error)
plt.ylabel('error')
plt.xlabel('datapoint')
print("the maximum error is",np.amax(error))
```
